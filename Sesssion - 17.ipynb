# Data Encapsulation

## Protected members

```python
class Base:
    def __init__(self):
        self._a = 2         #protected member of the class
        
class Derived (Base):
    def __init__(self):
        Base.__init__(self)
        print('Calling the protected member of the base class:', self._a)
        
        ## Modify the protected member of the base class
        self._a = 5
        print('Calling the modified member outside class:', self._a)
        
obj1 = Derived()
obj2 = Base()
```

```python
obj2._a
```

```python
obj1._a
```

## Private members

```python
class Base():
    def __init__(self):
        self.a = 'Edubridge'
        self.__c = 'Edubridge'
        
class Derived(Base):
    def __init__(self):
        Base.__init__(self)
        print('Calling the private member of the base class:', self.__c)
        
obj1 = Base()
```

```python
obj1.a
```

```python
obj1.c
```

```python
obj2 = Derived()
```

## Getter and Setter methods

```python
class Friend():
    def __init__(self):
        self.job = 'None'
        
    def getjob(self):
        return self.job
    
    def setjob(self, job):
        self.job = job
        
sam = Friend()
john = Friend()
```

```python
sam.setjob('Data Scientist')
john.setjob('SDE')
```

```python
print(sam.getjob())
```

```python
print(john.getjob())
```
