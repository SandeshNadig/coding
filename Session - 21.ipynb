```python
## Get the data for plotting sin and cos plots
X = np.linspace(-np.pi, np.pi, 256, endpoint = True)

## Assign the data to sin and cos
y, z = np.cos(X), np.sin(X)


plt.figure(figsize = (10, 6), dpi = 90)
plt.plot(X, y, color = 'pink', linewidth = 5, linestyle = '-.')
plt.plot(X, z, color = 'green', linewidth = 5, linestyle = ':')
```

```python
## Get the data for plotting sin and cos plots
X = np.linspace(-np.pi, np.pi, 256, endpoint = True)

## Assign the data to sin and cos
y, z = np.cos(X), np.sin(X)


plt.figure(figsize = (10, 6), dpi = 90)
plt.plot(X, y, color = 'pink', linewidth = 5, linestyle = '-', label = 'cosine')
plt.plot(X, z, color = 'green', linewidth = 5, linestyle = '-', label = 'sine')

plt.legend(loc = 'best')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

Z = np.random.uniform(0, 1, 20)

plt.figure(figsize = (15, 9))
plt.pie(Z)

plt.show()
```

```python
data = {'C' : 20, 'C++' : 15, 'Java' : 30, 'Python' : 35}

courses = list(data.keys())
numbers = list(data.values())

plt.bar(courses, numbers, color = 'green')
plt.xlabel('courses offered', fontsize = 22)
plt.ylabel('Number of students enrolled')
plt.show()
```

```python
plt.pie(numbers, labels = 'courses')
```
